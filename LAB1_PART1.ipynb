**LAB1: IMAGE AUGMENTATION & PREPARARTION**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("bird.jpg")
```

```python
cv2_imshow(image)
```

```python
from pyparsing.helpers import itemgetter
def image_augmentation(image, path):

  #fliping
  image_flip= cv2.flip(image,0)
  cv2.imwrite(path+"/birdflip.jpg",image_flip)
  #Cropping
  img_crop=image[10:100,50:150]
  cv2.imwrite(path+"/imageCrop.jpg",img_crop)
  #Resize
  img_resize=cv2.resize(image,(256,256))
  cv2.imwrite(path+"/resized.jpg",img_resize)
  #Rotation 
  h,w =image.shape[:2]
  M = cv2.getRotationMatrix2D((int(w/2),int(h/2)),45,1)
  img_rotation = cv2.warpAffine(image,M,(w,h))
  cv2.imwrite(path+"/img_rotation.jpg",img_rotation) 
  #Brightness
  matrix =np.ones(image.shape,dtype="uint8")*120
  img_Bright=cv2.add(image,matrix)
  cv2.imwrite(path+"/img_Bright.jpg", img_Bright)
  #MedianBlur
  img_MB=cv2.medianBlur(image,5)
  cv2.imwrite(path+"/img_MB.jpg ",img_MB)
  #Sharpening
  Kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
  img_sharpening = cv2.filter2D(image, -1, Kernel)
  cv2.imwrite(path+"/img_sharpening.jpg", img_sharpening)

  plt.subplot(2,4,1), plt.imshow(img_crop), plt.title("cropped")
  plt.subplot(2,4,2), plt.imshow(image_flip), plt.title("fliped")
  plt.subplot(2,4,3), plt.imshow(img_resize), plt.title("resized")
  plt.subplot(2,4,4), plt.imshow(img_rotation), plt.title("Rotation")
  plt.subplot(2,4,5), plt.imshow(img_Bright), plt.title("Brightness")
  plt.subplot(2,4,6), plt.imshow(img_MB), plt.title("MedianBlur")
  plt.subplot(2,4,7), plt.imshow(img_sharpening), plt.title("Sharpening")
 
```

```python
image_augmentation(image,'/content/augment')
```
